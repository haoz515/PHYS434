```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats
import pandas as pd

#this sets the size of the plot to something useful
plt.rcParams["figure.figsize"] = (5,5)
```

```python
# built op the data 
t = np.linspace(0,np.pi*100,np.int(1e5))
N = 10+np.sin(t)
bkgd = stats.norm.rvs(size=np.int(1e5))*np.sqrt(N)+N
```

```python
# define the function to do the slice
def slicing(x,y,start,size):
    end = start + size
    x_slice =  x[start:end]
    y_slice = y[start:end]
    plt.scatter(x_slice,y_slice,s = 0.2)
    plt.show()
```

```python
slicing(t, bkgd, 500, 2000)
```

```python
i = 0
new_t = [0] * int(1e5/100) 
new_bkgd = [0] * int(1e5/100) 

while i < 100000:
    index = int(i /100)
    new_t[index] = t[i]
    new_bkgd[index] = bkgd[i]
    i = i + 100
    
slicing(new_t,new_bkgd,0,int(1e5/100))
```

```python
plt.hist2d(t,bkgd,70)
plt.xlabel("time")
plt.ylabel("background")
plt.colorbar()
plt.show()
```

Brighter color means more data over there. Background is around 10. 

```python
t_period_2pi = np.mod(t,(np.pi*2))
plt.scatter(t_period_2pi,bkgd, s = 0.2)
plt.xlabel("time")
plt.ylabel("background")
plt.show()
```
